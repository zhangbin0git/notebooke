# 信息化评估的系统性分析如何实现信息化评估的全面性，做到没有死角

首先，不被陷入具体的系统中；其次，不被陷入具体的技术中；最后，不被陷入具体的部门（人员）中。  
系统性信息化评估所面临的挑战有：  
如何面对纷繁复杂的众多信息系统？  
如何不陷入具体的功能实现和操作？  
如何应对不断变化的管理职能和业务流程？  
如何在众多的评估中做到重点突出？  
如何反映系统的集成性、柔性和拓展性？

# 信息化评估的客观性分析
不带感情色彩、客观地对信息化现状进行评估，使得不同人进行评估，所出的结论保持基本一致，实践中往往因为以下因素的困扰，使得信息化评估的客观性很难得到保障。  
如何使得评估的结果不受管理者主观判断的干扰？  
如何使得评估的结果不受具体使用者喜好的干扰？  
如何使得评估的结果不受技术先进与否的干扰？（技术先进与否不代表信息化应用效果的优劣）  
如何使得评估的结果不受系统品牌的干扰？  


# 信息化评估的科学性分析
信息化评估的结果不是为了评估而评估，而是更好地指导企业进一步深化应用，凸显信息化的价值。  
如何直观、可视化地对评估结果进行展示，便于决策层、管理层进行有效的决策？  
如何使评估的结果有效地与企业的管理现状相结合，有效地推动企业的管理变革？  
如何一目了然地看到所处于的阶段，取得的成果、存在的差距？  
如何发现问题并提出深化应用的方向？如何能更好地看到前进的方向和目标？  

# 完整信息化应用评估包含由信息化整体评估和各应用系统评估、业务信息化评估、系统集成评估、数据评估、信息安全评估和信息管理评估组成的1+6的评估体系。

# 信息化整体评估：  
站在企业决策管理层的角度，以战略基调、以业务为核心，客观反映企业信息化建设中的总体情况，已经应用了哪些？应用状况如何？哪些还没有应用？在实践中，企业需要构建一个评估的标准体系，企业可根据《工业企业信息化和工业化融合评估规范》（GB/T23020-2013），也可借鉴本书中的模型构建符合企业实际需要的评估体系，在该体系的基础上组织评估。

# 应用系统评估：
从系统应用的视角，客观地评估该信息系统在企业中的应用情况，哪些功能进行了应用？哪些功能尚未应用？没有应用的原因如何？结合企业实际需求，
需求，哪些方面需要进行深化和拓展？


# 业务信息化评估：
站在企业管理的角度，以企业的管理业务为核心进行梳理，哪些管理业务已经通过信息化进行了覆盖，应用情况如何？如何进行深化应用？哪些业务尚未被信息化覆盖，是否需要实施相关的信息系统，实施的紧迫情况如何？

# 系统集成评估：
以信息系统集成的角度，客观评估现有哪些系统之间实现了集成？集成方式如何？哪些系统尚未实现集成，是否需要集成？集成的紧迫性如何？


## 数据评估：
站在数据的角度评估各应用系统的静态数据、动态数据的完备性、准确性和唯一性等，同时评估这些数据能否满足企业进行数据分析的要求？能否及时形成业务管理部门需要的报表？


# 信息安全评估：
以信息安全为核心，全面客观的对企业信息化的运行环境进行综合的评估，包含“硬环境”和“软环境”评估。硬环境的评估主要是支撑信息系统正常运行的硬件、网络及安全等硬件环境等；软环境评估主要包含人员管理与培训、配套制度的制定及执行情况等方面的评估。


# 信息管理评估：
以信息化实现的角度重点对信息化部门本身业务流程管理评估，如IT组织架构、信息化规划、需求管理、战略供方管理、实施管理、验收管理和升级维护管理等。根据不同的评估要求，信息化的评估工具有一定的区别，可形成信息化整体评估工具、应用系统评估工具、信息安全评估工具和信息管理评估工具等。

# 集团企业发展战略需求分析
对于集团企业而言，除了把握好集团企业的发展战略外，还需要理清集团的管控模式，**集团管控模式是集团企业制定未来发展的核心内容**。通常集团企业的信息化规划需要考虑几个层级信息化需求问题，首先是集团本部自身的信息化需求；其次是集团对下属企业的管控而产生的需求；最后可能需要根据业务板块来考虑板块间的信息化需求。要理清这些需求，核心是要理清集团的管控模式。**只有理清了管控模式，才能把握其真正的信息化需求，在此之上方能构架合理的信息化体系。**


# 所谓集团管控是指企业集团总部在管理下属企业中的定位，其具体体现在通过管控部门的设置、管控流程设计以及集团文化的传播来影响下属经营单位的战略、营销、财务、经营运作等方面的内容。通常的集团管控模式有财务管控、战略管控及运营管控等。
**财务管控型**：是指集团对下属子公司的管理控制主要通过财务手段来实现，集团对下属子公司的具体经营运作管理基本不加干涉，也不会对下属公司的战略发展方向进行限定，集团主要关注财务目标的实现，并根据业务发展状况增持股份或适时退出。  
**战略管控型**：是指集团的核心功能为资产管理和战略协调功能。集团与下属子公司的关系主要通过战略协调、控制和服务而建立，但是集团总部很少干预子公司的具体日常经营活动。集团根据外部环境和现有资源，制定集团整体发展战略，通过控制子公司的核心经营层，使子公司的业务活动服从于集团整体战略活动。一般地，这种情况比较适用于相关产业企业集团的发展。  
**运营管控型**：是指通过母公司的业务管理部门对控股子公司的日常经营运作进行直接管理，特别强调公司经营行为的统一、公司整体协调成长和对行业成功因素的集中控制与管理。这三种模式各具特点，运营控制型和财务控制型是集权和分权的两个极端，战略管控型则处于中间状态。有的企业集团从自己实际情况出发，为了便于管控，将处于中间状态的战略管控型进一步细化为“战略实施型”和“战略指导型”，前者偏重于集权而后者偏重于分权，但是仅有以上的模式分析是远远不能达到其信息化需求分析要求的。


# <font color='red'>注意：管控模式决定信息化架构</font>
作者曾经咨询过一家组建不久的集团企业，在采访其集团领导时，大家都一致认为本集团企业是运营管控型，并且因此提出了五统一的管理要求，即“统一经营、统一资金、统一采购、统一技术、统一协调”，但具体到某一个业务领域，应该管控到什么程度，其边界如何，就比较含糊不清。下面仅以统一采购为例说明，可以形成统一供应商目录、统一供应商寻源、统一供应商评价、统谈分签、统谈统签、统谈分签第三方集中配送、统谈统签第三方集中配送、统谈统签内部集中配送等多种形式，都可以叫作统一采购，但不同的模式对信息系统要求和IT构架是完全不一样的。
如果集团企业在进行信息化需求分析的时候不能很好地将管控关系分析清楚，就盲目拿出一个信息化框架，并按照“整体规划、分步实施”的思路付诸实施，其风险是非常大的，因为集团企业的信息化与单体企业的信息化建设有比较大的区别，不像单体企业，集团企业调整起来是非常困难的，有时候信息化实施过程是不可逆的。

# 关于核心系统的升级存在三种思路，即在维持现有的应用模式的基础上对系统进行升级；重新选型，实施更大型的信息系统；实施集成平台逐步实现升级，这三种模式的分析如下。

**维持现有的应用模式，对系统进行升级**：现有的IT人员和员工对现有系统比较熟悉，因此不会对现有的运作模式带来影响；资金投入较少；项目实施风险低，但是对企业的战略支撑相对较弱，对管理变革，业务流程优化提升等方面也相对较弱；系统间的集成、运维及升级存在较大的问题；更为关键的是现有的系统即便升级了，很多涉及业务要求的功能仍然无法很好满足，而且IT人员会陷入不断为业务部门进行二次开发、三次开发的循环中，陷入疲于奔命的礮境。  
**重新选型，实施大型信息系统**：可以很好的根据企业未来的发展来进行重新架构，能很好地支撑企业发展战略的实现；对推动企业变革管理等将起到积极的作用；大型信息系统往往功能更为强大，可以解决目前信息系统不能实现的功能；系统集成性相对较好。但是其一次性资金投入高；对IT人员和公司员工的素质要求较高；同时可能解决了部分问题，但也会带来新问题；项目的实施风险相对较大。  
**实施集成平台逐步实现升级**：通过实施集成平台逐步实现升级是融合了前面两种方案而选择的折中方案。一方面一方面很好地解决了系统之间的集成问题；另一方面也大大降低了选择大型信息系统一次性大投入的风险；更为关键的是即便选用大型信息系统，未来也需要考虑集成平台问题（是在大型系统内解决还是通用平台来进行解决）。当然选择集成平台逐步实现升级要对IT的实施思路进行较大的调整，尤其是对IT人员的素质要求更高，实施周期较长。

# 信息化集成战略分析
目前通用的信息系统集成模式主要有点对点集成、中心-辐条式集成和企业服务总线集成等三种模式，具体采取哪种方式，则需根据企业信息系统现状及对未来的期望进行选择，无论选择哪种集成模式，企业都需要根据选择的模式制定相应的规范。  
（1）点对点集成（见图7-3）企业中需要互通信息，共享数据的系统两两桥接起来，桥接的技术也是为两个特定系统专门定制通信链路来转换这两个系统的接口、协议以及数据格式等差异，通过这种点到点的简单连接，实现信息和数据的共享。点对点集成模型缺点：全部使用专门为两个特定系统定制的连接，在企业系统众多的情况下连接急剧增加，难以开发，后期维护及拓展困难。  
（2）中心-辐条（hub-spoke）式体系结构（见图7-4），该方式提供了一个应用集成中心（hub），这个中心具有自己的连接协议，所有需要集成的系统都和该中心相连。之前用户每集成一个系统，都要考虑该系统和其他所有系统的点对点连接的协议，数据结构的转换，而在中心-辐条结构下，用户只需要考虑系统和集成中心的点对点连接上转换。这样原来n个系统之间的n×（n-1）/2个点对点连接减少为n个连接。缺点：集中式的结构容易造成效率瓶颈，同时存在单点失效的问题。  
（3）企业服务总线（EnterpriseServiceBus，ESB）基于开放的标准消息总线，用于通过标准的适配器和接口来提供各程序和组件之间的互操作功能。它支持相互独立的异构环境中的服务、消息及基于事件的交互，并且具有适当的服务级别和可管理性。ESB通过开发面向应用的业务适配器组件，实现各集成应用之间可管理的接口透明，透明，为企业应用提供了便捷、一致、安全并符合标准的丰富接口，保证系统之间信息的可靠传送，进而实现不同厂商的数据库、应用系统之间的集成，如图7-5所示。缺点：技术比较复杂，对信息化人员的技能要求较高。


## 总之，企业在明晰了信息化组织战略、投资战略、集成战略、实施战略和绩效管理战略后，企业的信息化战略就变成了具体可操作的纲领，在此战略的指导下，根据企业发展战略、业务特点，结合管理和IT技术的最新发展趋势制定出相应的信息化整体规划，在整体规划的框架下组织信息化建设，可大大降低信息化实施的风险，实现各个信息系统的集成与融合，实现组织、业务流程与信息系统的融合，实现信息系统与自动化系统的融合，实现产品整个全生命周期信息的集成与融合，最终实现信息化战略对企业发展战略的传承。

# 信息化规划报告深度解析
第一节　规划报告需统一的思路
一、规划报告需遵循的原则企业进行信息化规划时通常应遵循的原则有：
战略优先原则：即在规划过程中将优先满足企业的发展战略，并以核心业务为纽带，实现以“战略+业务”为驱动的信息化规划。  
适用性原则：即在规划的过程中，充分考虑企业的信息化现状、行业特点以及未来的发展需求等条件，提出具备可操作性、针对性强的规划方案，能够指导企业进行规划方案的最终实现。  
集成性原则：即在规划的过程中，重点考虑跨业务、跨部门和跨系统之间的集成，在信息化建设开始就贯穿“体系”的思想，从整体考虑，避免出现信息孤岛。
适度超前原则：即在规划的过程中，考虑“先进管理模式+最新的IT技术”，在技术选择上，将采用适度超前的成熟技术，使得本轮规划在保持一定前瞻性的基础上能够付诸实施。  
拓展性原则：即在规划的过程中，将坚持规划的可扩展性，以满足企业快速发展的需求，同时兼顾系统的柔性，以满足企业动态发展的需要。
高性价比原则：即在规划中将考虑未来的信息化投入，重点以整体拥有成本（TOC）
二、规划报告需统一的概念信息化规划往往会涉及系统、平台、功能等概念，如果不明确好各自所表达的含义，在具体报告呈现的时候往往会相关概念交叉，平台中嵌套系统，系统中描述功能的现象，让管理者费解，因此需在规划之初，就需明确其含义，并在规划报告的组成过程中遵照执行。  
体系：泛指在一定范围内或同类事物间按照一定的秩序和内部联系组合而成的整体，例如，不同应用系统组成了信息化应用体系。  
平台：指一种基础的可用于衍生其他产品的环境，基于该环境可以产生其他的产品或应用。信息平台需具备相应的开发功能与执行逻辑，并可在此基础上构建应用。  
系统：为实现某一目标，同类事物按一定秩序和内部联系组合而成的集合体，例如，企业资源计划系统、人力资源管理系统等。  
模块：大型软件系统中能够单独命名并独立地完成一定功能的程序集合，例如，企业资源计划系统中的生产模块、采购模块等。  
三、规划报告所涵盖的内容  
信息化规划报告通常由需求分析、整体规划、系统设计和实施规划等四部分内容组成（见图8-1）。其中：  
需求分析：重点是前期需求分析报告通过讨论后的结果，是需求分析报告的摘要，此外加上最新的与本行业相关的管理发展和IT技术发展等方向的描述。  
整体规划：主要涵盖信息化SWOT分析、信息化战略目标、信息化蓝图设计和信息化战略分析等内容。  
系统设计：通常包含应用体系、支撑体系和保障体系等三大类核心内容的设计。应用体系设计：对核心应用系统进行初步设计，通常包含新建应用系统设计、已建应用系统深化和集成平台设计等内容，其中新建应用系统设计重点对系统目标、系统功能、系统间I/O分析等方面内容进行描述；支撑体系设计：主要对应用体系的支持基础环境进行设计，重点包括IT基础设施、信息安全和IT运维等方面的内容；保障体系设计：主要涵盖IT治理体系和组织构架设计等内容。  
实施规划：主要涵盖实施规划（实施原则、项目分解、优先级划分、实施规划）、投资分析等内容。
